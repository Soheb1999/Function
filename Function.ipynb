#Q.1. What is the difference between a function and a method in Python?

Ans:- In Python, both functions and methods are blocks of reusable code designed to perform specific tasks, but their key difference lies in their association with objects and classes.

* Function:
A function is a standalone block of code that is defined independently and can be called directly by its name from anywhere in the program. Functions typically operate on data passed to them as arguments and may or may not return a value. They are not tied to any specific object or class.

Example :-


```python
def greet(name):

    print(f"Hello, {name}!")

greet("Alice")
```

- Method:
A method is a function that is associated with an object or a class. It is defined within a class and operates on the data (attributes) of that specific object or class instance. Methods are called using dot notation on an object or class, and they implicitly receive the object instance (conventionally named self) as their first argument.

Example :-

```python
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        """This is a method that makes the dog bark."""
        print(f"{self.name} says Woof!")

my_dog = Dog("Buddy") # Creating an object (instance) of the Dog class
my_dog.bark() # Calling the method on the object
```

#Q.2. Explain the concept of function arguments and parameters in Python.

Ans :- In Python, the terms "parameters" and "arguments" are fundamental to understanding how functions receive and process data.

- Parameters are the variables defined within the parentheses of a function's definition. They act as placeholders for the values that the function is expected to receive when it is called. Parameters define the type and number of inputs a function anticipates.

- Arguments are the actual values or expressions passed to a function when it is called. These values are assigned to the corresponding parameters within the function's scope, allowing the function to operate on specific data during its execution.

Here is an example illustrating the concept:

```python
def greet(name, age):  # 'name' and 'age' are parameters
    print(f"Hello, {name}! You are {age} years old.")

# Calling the function with arguments
greet("Alice", 30)  # "Alice" and 30 are arguments
greet("Bob", 25)    # "Bob" and 25 are arguments
```

Ans :- The return statement in a Python function serves two primary purposes:

- Exiting the Function:
When a return statement is encountered within a function, the function's execution immediately terminates. Control is then transferred back to the point in the code where the function was called. Any code written after the return statement within that function will not be executed.

- Returning a Value (or Values):
The return statement can be used to send one or more values back to the caller of the function. This returned value can be stored in a variable, used in an expression, or passed as an argument to another function. If a function does not explicitly use a return statement, or if return is used without a specified value, the function implicitly returns None.

Example :-

```python
def calculate_area(length, width):
    area = length * width
    return area
rectangle_area = calculate_area(5, 10)
print(f"The area of the rectangle is: {rectangle_area}")

def greet(name):
    print(f"Hello, {name}!")
    return "Greeting complete."
message = greet("Alice")
print(message)
```

#Q.6.  Explain the concept of generators in Python and how they are defined.

Ans :- In Python, a generator is a special type of iterator that allows you to iterate over a potentially large sequence of values without storing them all in memory at once. This is achieved through the use of the yield keyword, which pauses the function's execution and returns a value, and resumes from where it left off when the next value is requested.

- Definition :
1. def keyword:
Generators are defined using the def keyword, just like regular functions.
2. yield keyword:
The key difference is the presence of the yield statement within the generator function.
3. Generator object:
When a generator function is called, it doesn't execute the code immediately. Instead, it returns a generator object.
4. Iteration:
You can iterate over this generator object using a for loop or the next() function. Each time next() is called (or the loop iterates), the generator function executes until it encounters another yield statement. The yielded value is then returned, and the generator's state is saved.
5. StopIteration:
When the generator function reaches the end (no more yield statements or returns), it raises a StopIteration exception, indicating that there are no more values to generate.
- Example :-

```python
def even_numbers(max_val):

  for num in range(2, max_val + 1, 2):
    yield num
even_gen = even_numbers(10)
for num in even_gen:
  print(num)
 print(next(even_gen))
 print(next(even_gen))
```

#Q.7.What are the advantages of using generators over regular functions?

Ans :- Generators offer several advantages over regular functions, particularly when dealing with sequences of data or iterative processes.
1. Memory Efficiency (Lazy Evaluation):
Generators compute and yield values one at a time, on demand, rather than generating and storing an entire sequence in memory upfront. This is crucial for handling large datasets or infinite sequences, preventing excessive memory consumption.

Example: A regular function to generate a large list of numbers would store all numbers in memory:

```python
    def generate_numbers_list(n):
        numbers = []
        for i in range(n):
            numbers.append(i)
        return numbers
```

2. Handling Infinite Sequences:
Due to their lazy evaluation, generators are ideal for representing and working with infinite sequences, as they only generate the next element when requested. Regular functions cannot practically handle infinite sequences as they would attempt to compute and store all elements, leading to memory exhaustion.

Example: Generating an infinite Fibonacci sequence:

```python
    def fibonacci_generator():
        a, b = 0, 1
        while True:
            yield a
            a, b = b, a + b
```

3. Simplified Iterative Logic:
Generators simplify the creation of iterators, eliminating the need for explicit __iter__ and __next__ methods required for custom iterator classes. The yield keyword handles the state management automatically, making the code more concise and readable.

Example: Creating a custom iterator for a sequence of squared numbers:

```python
    class Squares:
        def __init__(self, n):
            self.n = n
            self.current = 0

        def __iter__(self):
            return self

        def __next__(self):
            if self.current < self.n:
                result = self.current * self.current
                self.current += 1
                return result
            else:
                raise StopIteration
    def squares_generator(n):
        for i in range(n):
            yield i * i
```

#Q.8. What is a lambda function in Python and when is it typically used?

Ans :- In Python, a lambda function is a small, anonymous function defined without a name, using the lambda keyword. It can take any number of arguments but can only have one expression. Lambda functions are often used as inline functions within other functions, especially higher-order functions like map, filter, and reduce.

- Key characteristics of lambda functions:
- Anonymous: They don't have a name like regular functions defined with def.
- Single expression: They can only contain a single expression, which is evaluated and returned.
- Concise: They are useful for creating small, throwaway functions for specific tasks.

Example:

```python
def add(x, y):
    return x + y

add_lambda = lambda x, y: x + y

print(add(5, 3))
print(add_lambda(5, 3))
```

- Common use cases:

1.Higher-order functions: Lambda functions are frequently used as arguments to functions like map, filter, and reduce.

2.Inline functions: When you need a simple function for a short period and don't want to define a separate named function.

3.Data Manipulation with Pandas: In data science, lambda functions are often used with Pandas DataFrames to perform operations on Series (columns).

#Q.9.Explain the purpose and usage of the "map()" function in Python.

Ans :- The map() function in Python is a built-in function that provides a concise and often efficient way to apply a given function to each item of an iterable (like a list, tuple, or set) and return an iterator that yields the results.

- Purpose of map():
The primary purpose of map() is to transform elements of an iterable by applying a specified function to each one. It's a functional programming construct that allows you to perform operations on collections of data without explicitly writing a for loop, leading to more readable and sometimes more performant code.

Think of it as "mapping" a function over every item in a sequence.

Syntax of map():
The basic syntax of map() is:

```python
map(function, iterable, ...)
```

1. Basic Usage: Squaring Numbers

Let's say you have a list of numbers and you want to get a new list where each number is squared.

Using a for loop:

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = []
for num in numbers:
    squared_numbers.append(num * num)
print(squared_numbers)
```

Using map():

```python
numbers = [1, 2, 3, 4, 5]

def square(num):
    return num * num

squared_numbers_map = map(square, numbers)
print(squared_numbers_map)
result_list = list(squared_numbers_map)
print(result_list)
```

#Q.10. What is the difference between "map()", "reduce()", and "filter()" functions in Python?

Ans :- The map(), filter(), and reduce() functions in Python are built-in tools for functional programming, each serving a distinct purpose in processing iterables.
1. map():
Purpose: Applies a given function to each item in an iterable and returns a new iterable (a map object in Python 3) containing the results. It transforms each element independently.

Example:

```python
    numbers = [1, 2, 3, 4]
    squared_numbers = map(lambda x: x * x, numbers)
    print(list(squared_numbers)) # Output: [1, 4, 9, 16]
```

2. filter():

Purpose: Constructs a new iterable (a filter object in Python 3) containing only those items from the original iterable for which a given function returns True. It selects elements based on a condition.

Example:

```python
    numbers = [1, 2, 3, 4, 5, 6]
    even_numbers = filter(lambda x: x % 2 == 0, numbers)
    print(list(even_numbers)) # Output: [2, 4, 6]
```

3. reduce():

Purpose: Applies a given function cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value. It performs an aggregation. Note that reduce() is part of the functools module and needs to be imported.
Example:


```python
    from functools import reduce

    numbers = [1, 2, 3, 4]
    sum_of_numbers = reduce(lambda x, y: x + y, numbers)
    print(sum_of_numbers) # Output: 10
```

- Key Differences Summarized:
- Transformation vs. Selection vs. Aggregation:
map() transforms elements, filter() selects elements, and reduce() aggregates elements into a single result.

- Output Type:
map() and filter() return iterators (map and filter objects), while reduce() returns a single value.

- Number of Arguments for Function:
The function passed to map() and filter() typically takes one argument (an element from the iterable), while the function passed to reduce() takes two arguments (the accumulated result and the current element).

#Q.4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
import math

def is_prime(number):

    if number <= 1:
        return False
    if number == 2:
        return True
    if number % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(number)) + 1, 2):
        if number % i == 0:
            return False
    return True

def check_primes_up_to_200():

    print("Prime numbers from 1 to 200:")
    for num in range(1, 201):
        if is_prime(num):
            print(f"{num} is prime.")
        else:
            print(f"{num} is not prime.")
```

#Q.5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:

    def __init__(self, num_terms):

        if not isinstance(num_terms, int) or num_terms < 0:
            raise ValueError("num_terms must be a non-negative integer.")
        self._num_terms = num_terms
        self._current_term = 0
        self._a = 0
        self._b = 1

    def __iter__(self):

        return self

    def __next__(self):

        if self._current_term < self._num_terms:
            if self._current_term == 0:
                self._current_term += 1
                return self._a
            elif self._current_term == 1:
                self._current_term += 1
                return self._b
            else:
                next_fib = self._a + self._b
                self._a = self._b
                self._b = next_fib
                self._current_term += 1
                return next_fib
        else:
            raise StopIteration

fib_iter = FibonacciIterator(10)

print("Fibonacci sequence (first 10 terms):")
for num in fib_iter:
    print(num)

print("\nFibonacci sequence (first 5 terms using next()):")
fib_iter_2 = FibonacciIterator(5)
print(next(fib_iter_2))
print(next(fib_iter_2))
print(next(fib_iter_2))
print(next(fib_iter_2))
print(next(fib_iter_2))
```

#Q.6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two_generator(max_exponent):
    if not isinstance(max_exponent, int) or max_exponent < 0:
        raise ValueError("max_exponent must be a non-negative integer.")

    for i in range(max_exponent + 1):
        yield 2 ** i
```

#Q.7.  Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):

    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred while reading the file: {e}")

with open("example.txt", "w") as f:
    f.write("First line\n")
    f.write("Second line\n")
    f.write("Third line\n")
for line in read_file_line_by_line("example.txt"):
    print(line.strip())
```

#Q.9.  Write a Python program that uses "map()" to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(celsius):

  return (celsius * 9/5) + 32

def convert_temperatures(celsius_list):

    fahrenheit_list = list(map(celsius_to_fahrenheit, celsius_list))
    return fahrenheit_list

celsius_temps = [0, 10, 20, 30, 40]
fahrenheit_temps = convert_temperatures(celsius_temps)
print(f"Celsius temperatures: {celsius_temps}")
print(f"Fahrenheit temperatures: {fahrenheit_temps}")
```

#Q.10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_string):

    vowels = "aeiouAEIOU"

    def is_not_vowel(char):
        return char not in vowels

    filtered_chars = filter(is_not_vowel, input_string)

    return "".join(filtered_chars)

if __name__ == "__main__":
    test_string1 = "Hello World"
    result1 = remove_vowels(test_string1)
    print(f"Original string: '{test_string1}'")
    print(f"String without vowels: '{result1}'\n")

    test_string2 = "Programming is fun"
    result2 = remove_vowels(test_string2)
    print(f"Original string: '{test_string2}'")
    print(f"String without vowels: '{result2}'\n")

    test_string3 = "AEIOUaeiou"
    result3 = remove_vowels(test_string3)
    print(f"Original string: '{test_string3}'")
    print(f"String without vowels: '{result3}'\n")

    test_string4 = "Rhythm"
    result4 = remove_vowels(test_string4)
    print(f"Original string: '{test_string4}'")
    print(f"String without vowels: '{result4}'\n")
```
